```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
file_path = "C:\\Users\\swapn\\OneDrive\\Desktop\\Dev\\Zidio_Development_Internship\\dataset\\Online Retail.xlsx"

df = pd.read_excel(file_path)

print("Original shape:", df.shape)
df.head()
```

```python
print(df.info())

print("\nMissing values:")
print(df.isnull().sum())

print("\nDuplicate rows:")
print(df.duplicated().sum())
```

```python
df.columns = df.columns.str.strip()

df.rename(columns={
    "InvoiceNo": "invoice_no",
    "StockCode": "stock_code",
    "Description": "description",
    "Quantity": "quantity",
    "InvoiceDate": "invoice_date",
    "UnitPrice": "unit_price",
    "CustomerID": "customer_id",
    "Country": "country"
}, inplace=True)

df.head()
```

```python
df["invoice_no"] = df["invoice_no"].astype(str).str.strip()
df["stock_code"] = df["stock_code"].astype(str).str.strip()
df["description"] = df["description"].astype(str).str.strip()
df["country"] = df["country"].astype(str).str.strip()

df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
df["unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")
df["customer_id"] = pd.to_numeric(df["customer_id"], errors="coerce")

df["invoice_date"] = pd.to_datetime(
    df["invoice_date"],
    errors="coerce"
)

print(df.dtypes)
```

```python
print("Missing invoice_date:", df["invoice_date"].isnull().sum())
print("Date range:", df["invoice_date"].min(), "to", df["invoice_date"].max())

df[["invoice_no", "invoice_date"]].head()
```

```python
print("Duplicates before:", df.duplicated().sum())

df = df.drop_duplicates().copy()

print("Duplicates after:", df.duplicated().sum())
print("Shape after duplicate removal:", df.shape)
```

```python
df["description"] = df["description"].replace(["nan", "NaN", "", "None"], np.nan)
df["description"] = df["description"].fillna("Unknown Product")

df["customer_id_missing"] = df["customer_id"].isnull().astype(int)

print(df.isnull().sum())
```

```python
df["is_cancelled"] = df["invoice_no"].str.startswith("C").astype(int)

print(df["is_cancelled"].value_counts())

cancelled_df = df[df["is_cancelled"] == 1].copy()

cancelled_df.head()
```

```python
clean_df = df[df["is_cancelled"] == 0].copy()

clean_df = clean_df[
    (clean_df["quantity"] > 0) &
    (clean_df["unit_price"] > 0) &
    (clean_df["invoice_date"].notna())
].copy()

print("Clean sales shape:", clean_df.shape)
clean_df.head()
```

```python
clean_df["description"] = clean_df["description"].str.strip().str.lower()
clean_df["country"] = clean_df["country"].str.strip().str.title()
clean_df["stock_code"] = clean_df["stock_code"].str.strip()

clean_df.head()
```

```python
clean_df["total_amount"] = clean_df["quantity"] * clean_df["unit_price"]

clean_df[["quantity", "unit_price", "total_amount"]].describe()
```

```python
clean_df["date"] = clean_df["invoice_date"].dt.date
clean_df["year"] = clean_df["invoice_date"].dt.year
clean_df["month"] = clean_df["invoice_date"].dt.month
clean_df["day"] = clean_df["invoice_date"].dt.day
clean_df["day_of_week"] = clean_df["invoice_date"].dt.day_name()
clean_df["hour"] = clean_df["invoice_date"].dt.hour
clean_df["year_month"] = clean_df["invoice_date"].dt.to_period("M").astype(str)

clean_df.head()
```

```python
customer_df = clean_df.dropna(subset=["customer_id"]).copy()

customer_df["customer_id"] = customer_df["customer_id"].astype(int).astype(str)

print("Customer sales shape:", customer_df.shape)
customer_df.head()
```

```python
print("Clean sales shape:", clean_df.shape)
print("Customer sales shape:", customer_df.shape)
print("Cancelled orders shape:", cancelled_df.shape)

print("\nMissing values in clean_df:")
print(clean_df.isnull().sum())

print("\nDuplicate rows in clean_df:")
print(clean_df.duplicated().sum())

print("\nDate range:")
print(clean_df["invoice_date"].min(), "to", clean_df["invoice_date"].max())
```

```python
clean_df.to_csv("../data/processed/cleaned_online_retail_all_sales.csv", index=False)
customer_df.to_csv("../data/processed/cleaned_online_retail_customer_sales.csv", index=False)
cancelled_df.to_csv("../data/processed/cancelled_orders.csv", index=False)

print("Cleaned files saved successfully.")
```
