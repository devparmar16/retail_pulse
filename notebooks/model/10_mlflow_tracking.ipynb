```python
import mlflow

# Force MLflow to write into a consistent, local database file
mlflow.set_tracking_uri("sqlite:///mlflow.db")
mlflow.set_experiment("Retail_Analytics_Project")
print("Tracking URI set to SQLite database.")
```

```python
mlflow.set_experiment(
    "Retail_Analytics_Project"
)

print("Experiment Ready")
```

```python
with mlflow.start_run(run_name="Churn_Model"):

    mlflow.log_param(
        "model_type",
        "Blended XGBoost + RF"
    )

    mlflow.log_metric(
        "accuracy",
        0.7235
    )

    mlflow.log_metric(
        "precision",
        0.5534
    )

    mlflow.log_metric(
        "recall",
        0.8931
    )

    mlflow.log_metric(
        "f1_score",
        0.6834
    )

    mlflow.log_metric(
        "roc_auc",
        0.8265
    )

print("Churn Model Logged")
```

```python
with mlflow.start_run(run_name="Demand_Forecasting_Model"):

    mlflow.log_param(
        "model_type",
        "Ensemble XGBoost + RF"
    )

    mlflow.log_metric(
        "MAE",
        5133.68
    )

    mlflow.log_metric(
        "RMSE",
        7203.37
    )

    mlflow.log_metric(
        "SMAPE",
        19.97
    )

print("Forecast Model Logged")
```

```python
from mlflow.tracking import MlflowClient

client = MlflowClient()

exp = mlflow.get_experiment_by_name(
    "Retail_Analytics_Project"
)

runs = client.search_runs(
    experiment_ids=[exp.experiment_id]
)

print("Runs Found:", len(runs))

for run in runs:
    print(run.info.run_name)
```
